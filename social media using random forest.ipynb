```python
import pandas as pd
```

```python
data=pd.read_csv('mental_health_social_media_dataset.csv')
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
data.drop('person_name',axis=1,inplace=True)
```

```python
data.drop('date',axis=1,inplace=True)
```

```python
data.drop('positive_interactions_count',axis=1,inplace=True)
```

```python
data.drop('negative_interactions_count',axis=1,inplace=True)
```

```python
data.drop('platform',axis=1,inplace=True)
```

```python
data.drop('gender',axis=1,inplace=True)
```

```python
#data=pd.get_dummies(data,columns=['mental_state'],dtype='int')
```

```python
data.describe()
```

```python
X=data.drop('mental_state',axis=1)
y=data['mental_state']
```

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE()
X_sm, y_sm=sm.fit_resample(X,y)
```

```python
y_sm.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_ss=ss.fit_transform(X_sm)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test=train_test_split(X_ss,y_sm,test_size=0.2, random_state=20)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_validate
res=cross_validate(rf,X_train,y_train)
res
```

```python
print('validation score:', res['test_score'].mean())
```

```python
test_pred=rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy_score(y_test, test_pred)
```

```python
train_pred=rf.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy_score(y_train, train_pred)
```
