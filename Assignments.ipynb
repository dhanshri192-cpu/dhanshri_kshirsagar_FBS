```python
####q1]create a dataframe
#load the above dataset into a pandas dataframe.
```

```python
import pandas as pd
```

```python
data={'Order ID':[1001,1002,1003,1004,1005,1006,1007,1008],
      'Customer':['Alice','Bob','Charlie','Alice','David','Charlie','Bob','Alice'],
      'Product':['Laptop','Phone','Chair','Table','Laptop','Phone','Chair','Phone'],
      'Category':['Electronics','Electronics','Furniture','Furniture','Electronics','Electronics','Furniture','Electronics'],
      'Quantity':[2,1,4,1,1,2,2,1],
      'Price':[800,500,100,250,800,500,100,500],
      'Total Sales':[1600,500,400,250,800,1000,200,500]}
data1=pd.DataFrame(data)
data1
```

```python
###b]add a new column
##add a column discount that gives a 10% discount on total sales.
```

```python
data1['Discount']=data1['Total Sales']*0.10
data1
```

```python
###c] add new row
#add a new sales record :Order ID=1009,Customer='Eve',Product='Tablet',Category='Electronics',
#Quantity=1,Price=300,Total Sales=300.
```

```python
new_row={
    'Order ID':1009,
    'Customer':'Eve',
    'Product':'Electronics',
    'Category':1,
    'Quantity':1,
    'Price':300,
    'Total Sales':300,
    'Discount':30
}
data1=pd.concat([data1, pd.DataFrame([new_row])], ignore_index=True)
data1
```

```python
##filter data
#retrive only sales records where the category is electronics.

electronics_data=data1[data1['Category']=='Electronics']
electronics_data 
```

```python
# q5]selection
# select the customer and total sales 
```

```python
data=data1=data1[['Customer','Total Sales']]
data
```

```python
## 6]grouping (group the data by customer and calculate the total sales per customer)
```

```python
data=data1.groupby('Customer')['Total Sales'].sum()
data
```

```python
##7] sorting(sort the dataframe based on total sales in descending order
```

```python
data=data1.sort_values(by='Total Sales',ascending=False)
data
```

```python
##8]apply function
#apply a lambda function to create a new column sales category, where values are labeled as 'high' if total sales>1000,
#otherwise 'low'.
```

```python
data1['Sales Category']=data1['Total Sales']
data1
```

```python
##9] save dataframe
#same the final dataframe as a CSV file named sales_data.csv
```

```python
data1.to_csv('sales_data.csv',index=False)
```

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('dataset/tips.csv')
```

```python
data.head()
```

```python
## plot a day wise boxplot of totbill column
sns.boxplot(x='day',y='total_bill',data=data)
plt.title('day wise boxplt of total bill')
# plt.xlabel('day')
# plt.ylabel('total bill')
plt.show()
```

```python
##2]plot a pariplot by setting hue parameter t
```

```python
sns.pairplot(data,hue='sex')
plt.show()
```

```python
###3]draw a scatter plot of tip and totalbill and also allot the points according to the sex.
```

```python
sns.scatterplot(x='total_bill',y='tip', hue='sex', data=data)
plt.title('tip vs total bill')
plt.show()
```

```python
###4]plot histogram of totalbill for 50 bins
```

```python
plt.hist(data['total_bill'],bins=50)
plt.title('histogram of total_bill')
plt.xlabel('total_bill')
plt.ylabel('frequency')
plt.show()
```

```python
###plot a heatmap to represent entire dataset's correlation.
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation heatmap')
plt.show()
```
